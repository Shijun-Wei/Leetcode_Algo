# Sliding Window

```python
# Sliding Window 固定长度
def count_word(s, w):
    count = 0
    for i in range(len(s) - len(w) +1):
        if s[i:i+len(w)] == w.lower():
            count += 1
    return count
```

```python
# sliding window 固定k长度
def max_sum_of_k(nums, k):
	current_sum = sum(nums[:k])
	max_val = current_sum
	for i in range(len(nums)-k):
		current_sum = current_sum - nums[i] +nums[i+k]
		max_val = max(current_sum, max_val)
	return max_val
```

```python
# Sliding Window 不固定长度
def length_of_longest_substring(s):
    window_chars = set() # 记录窗口里的字符
    left = 0             # 屁股
    max_len = 0          # 最终答案
    # i 是头（右边界）
    for i in range(len(s)):
        while s[i] in window_chars:
            window_chars.remove(s[left])
            left += 1
        window_chars.add(s[i])
        max_len = max(max_len, i - left + 1)
        
    return max_len
```

```python
# Two pointer 快慢针
# 右指针 right：寻找“目标”（在这个题里是非零数）。
# 左指针 left：标记“存放目标的下一个位置”。
# two pointer approach： move zeros to the end of the array while maintaining the relative order of the non-zero elements.
# 当 right 找到一个非零数时，将其与 left 指向的位置交换，并将 left 向右移动一位，准备存放下一个非零数。
def move_zeros(nums):
    left, right = 0, 0
    while right  < len(nums):
        if nums[right] != 0:
            nums[left], nums[right] = nums[right], nums[left]
            left +=1
        right +=1
    return nums 
```

```python
# Two pointer 从两端往中间走 squre each element in a sorted array and return a new array of the squares in sorted order.
def sorted_squares(nums):
   n = len(nums)
   result = [0] * n
   left = 0
   right = n - 1
   for i in range(n - 1, -1, -1):
       if abs(nums[left]) > abs(nums[right]):
           result[i] = nums[left] * nums[left]
           left += 1
       else:
           result[i] = nums[right] * nums[right]
           right -= 1
   return result


def sorted_squares(nums):
    n = len(nums)
    # 准备一个结果数组，从后往前填
    res = [0] * n
    l, r = 0, n - 1
    pos = n - 1
    
    while l <= r:
        left_square = nums[l] ** 2
        right_square = nums[r] ** 2
        
        if left_square > right_square:
            res[pos] = left_square
            l += 1
        else:
            res[pos] = right_square
            r -= 1
        pos -= 1 # 填完一个，位置往前挪一位
        
    return res
```

```python
def merge_sorted_arrays(arr1, arr2):
    i, j = 0, 0
    merged = []
    
    # 当两个数组都还没走到头时
    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            merged.append(arr1[i])
            i += 1
        else:
            merged.append(arr2[j])
            j += 1
            
    # 将剩余的元素拼接到末尾（其中一个数组可能先跑完）
    merged.extend(arr1[i:])
    merged.extend(arr2[j:])
    
    return merged
```

```python
# binary search to count the number of negative numbers in a sorted 2D matrix (where each row is sorted in increasing order).
def count_negatives(matrix):
    count = 0
    n = len(matrix[0])
    
    for row in matrix:
        # 在每一行进行二分搜索
        l, r = 0, n - 1
        while l <= r:
            mid = (l + r) // 2
            if row[mid] < 0:
                l = mid + 1 # 负数在右边或当前位置
            else:
                r = mid - 1 # 正数，往左找负数
        # 循环结束时，l 就是这一行负数的个数
        count += l
    return count
```

```python
# Two pointer approach to count the number of negative numbers in a sorted 2D matrix (where each row is sorted in decending order, each column is sorted in descending order).
def count_negatives_optimal(grid):
    rows = len(grid)
    cols = len(grid[0])
    r, c = 0, cols - 1 # 从右上角开始
    count = 0
    
    while r < rows and c >= 0:
        if grid[r][c] < 0:
            # 如果右上角是负数，由于列也是有序的，
            # 那么这一位及它下方的所有数在该列都是负数
            count += (rows - r)
            c -= 1 # 往左看
        else:
            r += 1 # 太大了，往下面大数找
    return count
```

# Palindrome

```python
#palinfrome
def is_palindrome(s):
    left, right = 0, len(s) - 1
    while left < right:
        # 跳过非字母数字字符
        while left < right and not s[left].isalnum():
            left += 1
        while left < right and not s[right].isalnum():
            right -= 1
        
        # 比较字符（忽略大小写）
        if s[left].lower() != s[right].lower():
            return False
        
        left += 1
        right -= 1
    
    return True

# longest palindrome substring
def longest_palindrome(s):
    def expand_around_center(left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        return s[left + 1:right] # 返回当前回文子串

    longest = ''
    for i in range(len(s)):
        odd = expand_around_center(i,i)
        even = expand_around_center(i,i+1)
        longest = max(longest,even,odd, key= len)
    return longest

# dynamic programming approach to longest palindrome substring
def longest_palindrome(s):
    n = len(s)
    dp = [[False] * n for _ in range(n)]
    longest = ''
    
    for i in range(n):
        dp[i][i] = True
        longest = s[i]
    
    for length in range(2, n + 1): # 子串长度
        for start in range(n - length + 1):
            end = start + length - 1
            if s[start] == s[end]:
                if length == 2 or dp[start + 1][end - 1]: # 长度为2或内部也是回文
                    dp[start][end] = True
                    if length > len(longest):
                        longest = s[start:end + 1]
    
    return longest
```

```python
# Given a string, find the index of the first non-repeating character. If it doesn't exist, return -1.
def first_unique_character(s):
   char_count = {}
   for char in s:
       char_count[char] = char_count.get(char, 0) + 1
   for i, char in enumerate(s):
       if char_count[char] == 1:
           return i
   return -1
```

```python
#valid parentheses
#dfs approach
def is_valid_parentheses(s):
    stack = []
    # 建议将映射关系反过来，检查“左括号”可能更直观
    mapping = {")": "(", "}": "{", "]": "["}
    
    for char in s:
        if char in mapping:
            if not stack or stack.pop() != mapping[char]:
                return False
        else:
            stack.append(char)
            
    return not stack
```

```python
# invert a binary tree
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def invertTree(root):
    if not root:
        return None
    root.left, root.right = invertTree(root.right), invertTree(root.left)
    return root

# level order traversal of a binary tree
from collections import deque
def invertTree_bfs(root):
    if not root:
        return None
    
    # 1. 准备一个队列，先把根节点塞进去
    queue = deque([root])
    
    # 2. 只要队列里还有节点，就继续“清扫”
    while queue:
        # 3. 拿出当前最前面的那个节点
        node = queue.popleft()
        # 4. 核心：原地交换这个节点的左右手
        node.left, node.right = node.right, node.left
        # 5. 把它的左右孩子（如果有的话）放进队列，等待下一轮处理
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)      
    return root
```

```python
#leetcode 104 maximum depth of binary tree
def maxDepth(root):
    if not root:
        return 0
    left_depth = maxDepth(root.left)
    right_depth = maxDepth(root.right)
    return max(left_depth, right_depth) + 1
```

```python
#leetcode 543 diameter of binary tree
class Solution:
    def diameterOfBinaryTree(self, root):
        self.diameter = 0
        
        def depth(node):
            if not node:
                return 0
            
            left_depth = depth(node.left)
            right_depth = depth(node.right)
            
            # 更新全局直径
            self.diameter = max(self.diameter, left_depth + right_depth)
            
            # 返回当前节点的深度
            return max(left_depth, right_depth) + 1
        
        depth(root)
        return self.diameter
```

```python
#leetcode 110 balanced binary tree

class Solution:
    def isBalanced(self, root) -> bool:
        if not root:
            return True
        
        # 1. 计算左右子树的最大深度
        left_depth = self.maxDepth(root.left)
        right_depth = self.maxDepth(root.right)
        
        # 2. 判断当前节点是否平衡
        if abs(left_depth - right_depth) > 1:
            return False
        
        # 3. 递归检查左右子树
        return self.isBalanced(root.left) and self.isBalanced(root.right)

    # 补充这个函数
    def maxDepth(self, node):
        if not node:
            return 0
        return 1 + max(self.maxDepth(node.left), self.maxDepth(node.right))
```

```python
#leetcode 110 balanced binary tree
class Solution:
    def isBalanced(self, root) -> bool:
        def check_balance(node):
            if not node:
                return 0  # 空树的深度是0
            
            left_depth = check_balance(node.left)
            if left_depth == -1:  # 左子树不平衡 -1设置一个特殊值来表示不平衡
                return -1
            
            right_depth = check_balance(node.right)
            if right_depth == -1:  # 右子树不平衡
                return -1
            
            if abs(left_depth - right_depth) > 1:  # 当前节点不平衡
                return -1
            
            return max(left_depth, right_depth) + 1  # 返回当前节点的深度
        
        return check_balance(root) != -1
```
